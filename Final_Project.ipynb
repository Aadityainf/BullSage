```python
col=stock_data.columns.drop(['Ticker','Date'])
col
```

```python
for val in col:
    Q1 = stock_data[val].quantile(0.25)
    Q3 = stock_data[val].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = [x for x in stock_data[val] if x < lower or x > upper]
    print(val)
    print('Identified outliers: %d' % len(outliers))
    print('*'*40)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = stock_data['Close'].quantile(0.25)
Q3 = stock_data['Close'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows that are outliers
df_no_outliers = stock_data[(stock_data['Close'] >= lower_bound) & (stock_data['Close'] <= upper_bound)]

print(f"Data after removing outliers using IQR method: {df_no_outliers.shape}")
```

```python
# Check for any remaining outliers after removing them
print(df_no_outliers.describe())  # Check the summary statistics

# You can also visualize the distribution again (optional)
import matplotlib.pyplot as plt
plt.boxplot(df_no_outliers['Close'], vert=False)
plt.title('Box Plot after Removing Outliers')
plt.show()
```
